```python
import pandas as pd
year = 2008
voc = 'ethene'
df_year = pd.read_excel('LMR_VOCdata_97-19_DOW.xls',sheet_name=str(year))

VOCs = ['Ozone', 'Nitric oxide', 'Nitrogen dioxide',
       'Sulphur dioxide', 'Carbon monoxide', 'ethane', 'ethene', 'ethyne',
       'propane', 'propene', 'iso-butane', 'n-butane', '1-butene',
       'trans-2-butene', 'cis-2-butene', 'iso-pentane', 'n-pentane',
       '1,3-butadiene', 'trans-2-pentene', 'cis-2-pentene (VOC-AIR only)',
       '2-methylpentane', '3-methylpentane (VOC-AIR only)', 'isoprene',
       'n-hexane', 'n-heptane', 'benzene', 'toluene', 'ethylbenzene',
       'm+p-xylene', 'o-xylene', '1,2,4-trimethylbenzene',
       '1,3,5-trimethylbenzene']

for col in VOCs:
    try:
        df_year[col] = pd.to_numeric(df_year[col], errors='coerce')
    except:
        pass
df_year['Time'] = pd.to_datetime(df_year['Time'], format='%H:%M:%S',errors='coerce').dt.hour

mondays = df_year.loc[df_year['DOW'] == 'Monday']
mondays.groupby('Time', as_index=False, sort=False)[voc].mean()

tuesdays = df_year.loc[df_year['DOW'] == 'Tuesday']
tuesdays.groupby('Time', as_index=False, sort=False)[voc].mean()

wednesdays = df_year.loc[df_year['DOW'] == 'Wednesday']
wednesdays.groupby('Time', as_index=False, sort=False)[voc].mean()

thursdays = df_year.loc[df_year['DOW'] == 'Thursday']
thursdays.groupby('Time', as_index=False, sort=False)[voc].mean()

fridays = df_year.loc[df_year['DOW'] == 'Friday']
fridays.groupby('Time', as_index=False, sort=False)[voc].mean()

saturdays = df_year.loc[df_year['DOW'] == 'Saturday']
saturdays.groupby('Time', as_index=False, sort=False)[voc].mean()

sundays = df_year.loc[df_year['DOW'] == 'Sunday']
sundays.groupby('Time', as_index=False, sort=False)[voc].mean()
```
